```python
# @title 1. Setup Environment
# Clone the repository
!git clone https://github.com/shimoda-ryu/google-sheet-color-analyzer.git
%cd google-sheet-color-analyzer

# Switch to the feature branch with rembg support (Optional, remove if merged to main)
!git checkout feature/rembg-support

# Install Dependencies
!pip install -r requirements.txt

# Create .env file template
import os
if not os.path.exists('.env'):
    with open('.env', 'w') as f:
        f.write("")
```

```python
# @title 2. Configuration
SPREADSHEET_ID = "YOUR_SPREADSHEET_ID_HERE" # @param {type:"string"}
CREDENTIALS_FILE = "credentials.json" # @param {type:"string"}

import os

# Validate
if SPREADSHEET_ID == "YOUR_SPREADSHEET_ID_HERE":
    print("‚ö†Ô∏è Please enter your Spreadsheet ID above!")
elif not os.path.exists(f"../{CREDENTIALS_FILE}") and not os.path.exists(CREDENTIALS_FILE):
    print(f"‚ö†Ô∏è Please upload {CREDENTIALS_FILE} to the file browser (left icon)!")
else:
    # Check where credentials is
    creds_path = CREDENTIALS_FILE
    if os.path.exists(f"../{CREDENTIALS_FILE}"):
        creds_path = f"../{CREDENTIALS_FILE}"
    
    # Write to .env
    with open(".env", "w") as f:
        f.write(f"SPREADSHEET_ID={SPREADSHEET_ID}\n")
        f.write(f"GOOGLE_APPLICATION_CREDENTIALS={os.path.abspath(creds_path)}\n")
    
    print("‚úÖ Configuration saved!")
```
